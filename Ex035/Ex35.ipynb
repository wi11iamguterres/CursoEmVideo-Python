```python
comp1 = input("Digite o comprimento do primeiro lado: ")
comp2 = input("Digite o comprimento do segundo lado: ")
comp3 = input("Digite o comprimento do terceiro lado: ")


if comp1<comp2+comp3 and comp2<comp1+comp3 and comp3<comp1+comp2:
    print("É um triângulo.")
else:
    print("Não é um triângulo.")
```
