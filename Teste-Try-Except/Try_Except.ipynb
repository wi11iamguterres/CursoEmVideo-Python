```python
def pedir_nome():
  while True:
    try:
      nome=input("Digite seu nome: ").strip()
      if not nome:
        raise ValueError("O nome não pode estar vazio.")
      if nome.isdigit():
        raise ValueError("O nome não pode ser um número.")
      print("Olá, {}!".format(nome))
      break
    except ValueError as erro:
      print("Erro: {}.".format(erro))

pedir_nome()
```
