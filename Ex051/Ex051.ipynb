Desenvolva um programa que leia o primeiro termo e a razão de uma PA. No final, mostre os 10 primeiro termos desse progressão.

```python
primeiroTermo = int(input('Digite o primeiro termo da PA: '))
razao = int(input('Digite a razão da PA: '))
decimoTermo = primeiroTermo + (10 - 1) * razao

for i in range(primeiroTermo, decimoTermo + razao, razao):
  print('{}'.format(i))
```
