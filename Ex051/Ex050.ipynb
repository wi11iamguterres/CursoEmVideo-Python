Desenvolva um programa que leia 6 números inteiros e mostre a soma somente daqueles que forem pares. Se o valor digitado for ímpar, desconsidere-o.

```python
7
soma=0
for n in range(1,7):
  num=int(input('Informe o {}º número: '.format(n)))
  if num % 2 == 0:
    soma=soma+num
print('A soma dos números pares é: {}'.format(soma))
```
