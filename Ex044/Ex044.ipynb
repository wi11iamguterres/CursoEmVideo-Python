Exercício Python 44: Elabore um programa que calcule o valor a ser pago por um produto, considerando o seu preço normal e condição de pagamento:

– à vista dinheiro/cheque: 10% de desconto

– à vista no cartão: 5% de desconto

– em até 2x no cartão: preço formal

– 3x ou mais no cartão: 20% de juros

```python

preco = float(input('Digite o valor do produto: '))
quantidade=int(input('Digite a quantidade de produto(s): '))
opcao=int(input(('''Escolha a forma de pagamento:
                [1] = à vista/dinheiro:
                [2] = à vista no cartão:
                [3] = em até 2x no cartão:
                [4] = 3x ou mais no cartão:\n
FORMA DE PAGAMENTO SELECIONADA: ''')))

if opcao == 1:
  print('''
  ===========PAGAMENTO À VISTA==========
  Valor do produto: R${:.2f}.
  Valor com 10% de desconto: R${:.2f}.
  Quantidade:{}.
  Valor total: R${:.2f}.'''.format(preco, preco*0.9, quantidade, ((preco*0.9)*quantidade)))
elif opcao == 2:
  print('''
  ===========PAGAMENTO À VISTA NO CARTÃO==========
  Valor do produto: R${:.2f}.
  Valor com 5% de desconto: R${:.2f}.
  Quantidade:{}.
  Valor total: R${:.2f}.'''.format(preco, preco*0.95, quantidade,((preco*0.95)*quantidade)))
elif opcao == 3:
  print('''
  ===========PAGAMENTO EM ATÉ 2X NO CARTÃO==========
  Valor do produto: R${:.2f}.
  Quantidade:{}.
  Valor da parcela: R${:.2f}.
  Valor total: R${:.2f}.'''.format(preco, quantidade, (preco*quantidade)/2, preco*quantidade))
elif opcao == 4:
  print('\n===========PAGAMENTO 3X OU MAIS NO CARTÃO==========')
  parcela=int(input('Digite a quantidade de parcelas: '))
  while parcela<3:
    print('Quantidade de parcelas insuficiente.')
    parcela=int(input('Digite a quantidade de parcelas: '))
  else:
    print('''
  Valor do produto: R${:.2f}.
  Valor com 20% de juros: R${:.2f}.
  Quantidade:{}.
  Valor da parcela: R${:.2f}.
  Valor total: R${:.2f}.'''.format(preco, preco*1.2, quantidade, (((preco*1.2)*quantidade)/parcela),((preco*1.2)*quantidade)))
else:
  print('Opção inválida.')
```
