Desenvolva um programa que leia o nome, idade e sexo de 4 pessoas. No final do programa mostre:
- A média de idade do grupo;
- Qual é o nome do homem mais velho.
- Quantas mulheres têm menos de 20 anos.

```python
soma=0
maisVelho=0
menorDe20=0
for i in range(1,5):
  nome=str(input('Digite o nome da {}° pessoa: '.format(i)).strip())
  idade=int(input('Digite a idade da {}° pessoa: '.format(i)).strip())
  soma=soma+idade
  sexo=str(input('Digite o sexo da {}° pessoa: '.format(i)).strip())
  if sexo=='M':
    if idade>maisVelho:
      maisVelho=idade
      nomeMaisVelho=nome
  if sexo=='F':
    if idade<20:
      menorDe20+=1

media=soma/4

print('\n=========RESULTADOS=========')
print('A média de idade do grupo é de {} anos.'.format(media))
print('O homem mais velho tem {} anos e se chama {}.'.format(maisVelho,nomeMaisVelho))
print('Existem {} mulheres com menos de 20 anos.'.format(menorDe20))
```
