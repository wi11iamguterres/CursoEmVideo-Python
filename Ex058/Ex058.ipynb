Melhore o jogo do DESAFIO 028 onde o computador vai "pensar" em um número entre 0 e 10. Só que agora o jogador vai tentar adivinhar até acertar, mostrando no quantos palpites foram necessários para vencer.

```python
from random import randint
import time
tentativa=0
jogador=int(input('Digite um número de 0 a 10: '))
tentativa += 1
computador = randint(0, 10)
print('Pensando.')
time.sleep(1)
print('Pensando..')
time.sleep(1)
print('Pensando...')
time.sleep(1)
while jogador != computador:
  time.sleep(1)
  print('Errou!')
  jogador=int(input('Digite um número de 0 a 10: '))
  print('Pensando.')
  time.sleep(1)
  print('Pensando..')
  time.sleep(1)
  print('Pensando...')
  time.sleep(1)
  tentativa += 1
else:
  print('Você acertou! Para isso foram necessárias {} tentativas.'.format(tentativa))


```
