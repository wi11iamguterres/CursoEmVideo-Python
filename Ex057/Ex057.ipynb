Faça um programa que leia o sexo de uma pessoa, mas só aceite os 'M' ou 'F'.
Caso esteja errado, peça a digitação novamente até ter um vaalor correto.

```python
sexo=str(input('Informe seu sexo [M/F]: ')).strip().upper()
while sexo != 'M' and sexo != 'F':
  print('Dados inválidos!')
  sexo=str(input('Informe seu sexo [M/F]: ')).strip().upper()

print('Sexo {} registrado com sucesso!'.format(sexo))
```
