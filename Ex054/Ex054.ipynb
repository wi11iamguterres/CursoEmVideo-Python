Crie um programa qua leia o anos de nascimento de sete pessoas. No final, mostre quantas pessos ainda não atingiram a maioridade quantas já são maiores.

```python
from datetime import date
contMais18=0
contMenos18=0
for a in range(1,8):
  ano=int(input('Informe o ano de nascimento da {}º pessoa: '.format(a)))
  if date.today().year - ano>=18:
    contMais18+=1
  else:
    contMenos18+=1
if contMais18==1:
  print('{} pessoa já atingiu a maioridade.'.format(contMais18))
elif contMais18==0:
  print('Nenhuma pessoa atingiu a maioridade.')
else:
  print('{} pessoas já atingiram a maioridade.'.format(contMais18))
if contMenos18==1:
  print('{} pessoa ainda não atingiu a maioridade.'.format(contMenos18))
elif contMenos18==0:
  print('Nenhuma pessoa atingiu a maioridade.')
else:
  print('{} pessoas ainda não atingiram a maioridade.'.format(contMenos18))
```
