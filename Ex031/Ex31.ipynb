```python
while True:
  try:
    distancia=float(input("Digite a distância em Km da viagem a ser percorrida: "))
    if distancia < 0:
      print("Distância inválida.")
    if distancia > 0 or distancia <= 200:
      print("Para uma viagem de {:.2f}Km a passagem custará R${:.2f}.".format(distancia,distancia*0.5))
      break
    else:
      print("Para uma viagem de {:.2f}Km a passagem custará R${:.2f}.".format(distancia,distancia*0.45))
      break
  except ValueError:
    print("Caráctere inválido.")
```
