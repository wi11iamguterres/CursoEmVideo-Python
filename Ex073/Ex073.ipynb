Crie uma tupla preenchida com os 20 primeiros colocados da Tabela do Campeonato Brasileiro de Futebol, na ordem da colocação.Depois mostre:
A)Apenas os 5 primeiros colocados;
B)Os últimos 4 colocados da tabela;
C)Uma lista com os times em ordem alfabética;
D)Em que posição na tabela está o time da Chapecoense;

```python
times = ('Flamengo','Palmeiras','Cruzeiro','Mirassol','Fluminense',
         'Botafogo','Bahia','São Paulo','Grêmio','Bragantino',
         'Atlético-MG','Santos','Corinthians','Vasco da Gama',
         'Vitória','Internacional','Ceará SC','Fortaleza',
         'Juventude','Sport Recife')
print('='*100)
print('A)')
print(f'Os 5 primeiros colocados foram: {times[0:5]}')
print('='*100)
print('B)')
print(f'Os últimos 4 colocados foram: {times[16:]}')
print('='*100)
print('C)')
print('A lista dos times por ordem alfabetica: ',sorted(times))
print('='*100)
print('D)')
print('A equipe do Santos ficou na',times.index('Santos')+1,'º colocação.')
print('='*100)
```
