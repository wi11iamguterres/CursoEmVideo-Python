Exercício Python 45: Crie um programa que faça o computador jogar Jokenpô com você.

```python
from random import randint
jogador=int(input('Escolha sua jogada: 1-Pedra, 2-Papel, 3-Tesoura: '))
comp=randint(1,3)
if jogador==comp:
  print('Empate. Você e o computador escolheram {}.'.format(comp))
elif jogador==1 and comp==2:
  print('VOCÊ PERDEU! Computador escolheu {}. Papel enrola Pedra.'.format(comp))
elif jogador==1 and comp==3:
  print('VOCÊ VENCEU! Computador escolheu {}. Pedra afia Tesoura.'.format(comp))
elif jogador==2 and comp==1:
  print('VOCÊ VENCEU! Computador escolheu {}. Papel enrola Pedra.'.format(comp))
elif jogador==2 and comp==3:
  print('VOCÊ PERDEU! Computador escolheu {}. Tesoura corta Papel.'.format(comp))
elif jogador==3 and comp==1:
  print('VOCÊ PERDEU! Computador escolheu {}. Pedra afia Tesoura.'.format(comp))
elif jogador==3 and comp==2:
  print('VOCÊ VENCEU! Computador escolheu {}. Tesoura corta Papel.'.format(comp))
else:
  print('Jogada Inválida!')
```
