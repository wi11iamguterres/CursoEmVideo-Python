Crie um programa que tenha uma tupla única com nomes de produtos e seus respectivo preços na sequencia.
No final, mostre uma listagem de preços, organizando os dados em forma tabular.

```python
produto=('Refrigerante', 9.58,
         'Água', 2.99,
         'Vazio', 59.90,
         'Contrafile',69.90,
         'Salsichão',19.90,
         'Cerveja',4.99,
         'Pão de Alho',15.90)

print('='*50)
print('TABELA DE PREÇOS')
print('='*50)
print('PRODUTO')
for pos in range(0, len(produto)):
  print(f'{produto[pos]:.<20}')

```
