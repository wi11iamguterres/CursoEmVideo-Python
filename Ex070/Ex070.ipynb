Crie um programa que leia o nome e o preço de vários produtos. O programa deverá perguntar se o usuário vai continuar. No final, mostre:
A) Qual o total gasto na compra; B)Quantos produtos custam mais de R$1000; C)Qual o nome do produto mais barato;

```python
total = 0
cont_mil = 0 # Counter for products above R$1000
menor_preco = float('inf') # Initialize with positive infinity to ensure the first product's price is smaller
nome_mais_barato = '' # Initialize with an empty string

print('-' * 50)
print('LOJA SUPER BARATÃO')
print('-' * 50)
while True:
  nome = input('Digite o nome do produto: ')
  while not nome.isalpha():
    nome = input('Nome inválido! Por favor, digite um nome válido.: ')
  else:
    while True:
        try:
            preco = float(input('Digite o valor do produto: '))
            if preco < 0:
                print('O preço não pode ser negativo. Tente novamente.')
                continue
            break
        except ValueError:
            print('Valor inválido. Por favor, digite um número para o preço.')

    total += preco # Accumulate total spent

    if preco > 1000:
        cont_mil += 1 # Increment count for products over R$1000

    # Correct logic for finding the cheapest product and its name
    if preco < menor_preco:
        menor_preco = preco
        nome_mais_barato = nome

    while True:
        continua = input('Deseja continuar [S] ou [N]? ').upper().strip()[0]
        if continua in ['S', 'N']:
            break
        else:
            print('Resposta inválida. Por favor, digite S para continuar ou N para parar.')

    if continua == 'N':
        print('FIM.')
        break # Exit the loop if the user does not want to continue

print('Total gasto: R${:.2f}'.format(total))
print('Produtos acima de R$1000,00: {}'.format(cont_mil))
print('Produto mais barato: {}'.format(nome_mais_barato))
```
