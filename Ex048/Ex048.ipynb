Faça um programa que calcule a soma entre todos os números impares que são múltiplos de três e que se encontra no intervalo de 1 a 500.

```python
soma=0
for i in range(1,501):
  if i % 2 != 0 and i % 3 == 0:
    soma=soma+i
print('A somo dos números ímpares que são múltiplos de 3 é: {}'.format(soma))

#Forma Guanabara:
'''soma = 0
cont = 0
for c in range(1, 501, 2):
  if c % 3 == 0:
    soma += c
    cont += 1
print('A soma dos {} números ímpares que são múltiplos de 3 no intervalo 1 a 500 é {}.'.format(cont,soma))'''
```
