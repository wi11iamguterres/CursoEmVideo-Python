Faça um programa que jogue par ou ímpar com o computador. O jogo só será interrompido quando o jogador perder, mostrando total de vitórias consecutivas que ele conquistou no final.

```python
from random import randint
cont=0
while True:
  jog=int(input('Escolha 2 [PAR] ou 1 [ÍMPAR]: '))
  while jog != 1 and jog != 2:
    print('Escolha inválida!')
    jog=int(input('Escolha 2 [PAR] ou 1 [ÍMPAR]: '))
  comp=randint(1,2)
  if comp==1 and jog==2:
    print('Você perdeu!')
    break
  elif comp==2 and jog==1:
    print('Você perdeu!')
    break
  else:
    print('Você ganhou!')
    cont+=1
print('Você ganhou {} vezes!'.format(cont))



```
