```python

frase = str(input("Digite uma frase: ")).strip().upper()
print("Há {} letra(s) 'A' na frase inserida.".format(frase.count("A")))
print("A primeira posição em que aparece a letra 'A' é na posição {}.".format(frase.find("A")+1))
print("A última posição em que aparece a letra 'A' é na posição {}.".format(frase.rfind("A")+1))
```
