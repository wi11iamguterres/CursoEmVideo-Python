Crie um programa que leia a idade e o sexo de vários pessoas. A cada pessoa cadastrada, o programa deverá perguntar se o usuário quer ou não continuar. No final, mostra: A)Quantas pessoas tem mais de 18 anos; B) Quantos homens foram cadastrados; C) Quantas mulheres tem menos de 20 anos;

```python
contIdade = contHomens = contMulheres = 0

while True:
  print('='*50)
  print('TABELA DE PESSOA')
  print('='*50)
  idade = int(input('Digite a idade: '))
  if idade>=18:
    contIdade+=1
  sexo = str(input('Digite o sexo: ')).upper()
  if sexo == 'M':
    contHomens+=1
  else:
    if idade<20 and sexo == 'F':
      contMulheres+=1
  continuar = str(input('Deseja continuar? [S/N]')).upper()
  if continuar == 'N':
    break
print('\nDados inseridos:')
print('Número de pessoas com mais de 18 anos: {}'.format(contIdade))
print('Número de homens cadastrados: {}'.format(contHomens))
print('Número de mulheres com menos de 20 anos: {}'.format(contMulheres))

```
