Faça um programa que leia o peso de cinco pessoas. No final, mostre qual foi o maior e o menor peso lidos.

```python
listaPeso=[]
for i in range(1,6):
  peso=float(input('Digite o peso: '))
  while peso<=0:
    print('Digite um peso válido!')
    peso=float(input('Digite o peso: '))
  else:
    listaPeso.append(peso)

print('\n==========RESULTADOS==========')
print('O maior peso é: {:.2f}Kg'.format(max(listaPeso)))
print('O menor peso é: {:.2f}Kg'.format(min(listaPeso)))
```
