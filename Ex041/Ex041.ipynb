Exercício Python 041: A Confederação Nacional de Natação precisa de um programa que leia o ano de nascimento de um atleta e mostre sua categoria, de acordo com a idade:

– Até 9 anos: MIRIM

– Até 14 anos: INFANTIL

– Até 19 anos: JÚNIOR

– Até 25 anos: SÊNIOR

– Acima de 25 anos: MASTER

```python
from datetime import date
ano=int(input('Digite o ano de nascimento do atleta: '))
idade=date.today().year-ano

if idade<=0:
  print('Idade inválida.')
elif idade>0 and idade<=9:
  print('Ano de nascimento é {}. Com {} anos o atleta pertence a categoria MIRIM.'.format(ano,idade))
elif idade<=14:
  print('Ano de nascimento é {}. Com {} anos o atleta pertence a categoria INFANTIL.'.format(ano,idade))
elif idade<=19:
  print('Ano de nascimento é {}. Com {} anos o atleta pertence a categoria JÚNIOR.'.format(ano,idade))
elif idade<=25:
  print('Ano de nascimento é {}. Com {} anos o atleta pertence a categoria SÊNIOR.'.format(ano,idade))
else:
  print('Ano de nascimento é {}. Com {} anos o atleta pertence a categoria MASTER.'.format(ano,idade))
```
